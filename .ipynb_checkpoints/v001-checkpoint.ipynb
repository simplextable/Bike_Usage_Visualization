```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
bike_demand = pd.read_csv("2011_bike_sharing.csv")
weather_condition = pd.read_csv("weather_condition.csv")
```

```python
weather_condition["Date"] = pd.to_datetime(weather_condition["DATE"])
```

```python
weather_condition["Common_Date"] = weather_condition["Date"]
```

```python
weather_condition
```

```python
weather_condition.drop(["STATION", "STATION_NAME", "DATE", "date2","Date"],axis=1,inplace=True) 
```

```python
bike_demand
```

```python
bike_demand["Start date"] = pd.to_datetime(bike_demand["Start date"])
```

```python
bike_demand['Start date2'] = bike_demand['Start date'].dt.floor('h')
```

```python
bike_demand["Common_Date"] = bike_demand["Start date2"]
```

```python
bike_demand.drop(["Start station number", "Start station", "End station number", "End station","Bike number","Start date2"],axis=1,inplace=True) 
```

```python
common_data = bike_demand.merge(weather_condition, how='left', on='Common_Date')
```

```python
common_data.isnull().sum()
```

```python
common_data = common_data.dropna()
```

```python
common_data.HPCP.unique()
```

```python
common_data
```

```python
common_data.to_csv('common_data.csv',index=False)
```
